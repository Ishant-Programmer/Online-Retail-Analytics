#### Observation
Negative quantities are not always associated with cancelled invoices.
Approximately **3,457 records** contain negative quantities without a corresponding invoice starting with **'C'**.
A closer inspection shows that these records commonly have:
- Price equal to **0**
- Missing **CustomerID**
- Inventory adjustment descriptions such as **"missing"**, **"smashed"**, **"check"**, etc.
These records appear to represent internal inventory adjustments rather than customer sales transactions.

#### Cleaning the data

```python
# Removing exact duplicate rows
df= df.drop_duplicates()
```

```python
# Verifying
df.duplicated().sum()
```

```python
# Removing Cancelled orders (Invoices starting with C)
df=df[~df["Invoice"].astype(str).str.startswith("C")]
```

```python
# Verifying
df["Invoice"].astype(str).str.startswith("C").sum()
```

```python
# Removing Negative Quantities
df=df[df["Quantity"]>0]
```

```python
# Verifying
(df["Quantity"]<0).sum()
```

```python
# Removing Negative Price
df=df[df["Price"]>0]
```

```python
# Verifying
(df["Price"]<0).sum()
```

```python
# Removing records with no description
df=df.dropna(subset=["Description"])
```

```python
# Verifying
df["Description"].isnull().sum()
```

Customer ID will be retained for now missing Customer IDs will be handled separately during customer analysis

#### Feature Engineering

```python
# Creating Revenue Column
df["Revenue"]=df["Quantity"]*df["Price"]
```

```python
# Converting InvoiceDate to datetime
df["InvoiceDate"]=pd.to_datetime(df["InvoiceDate"])
```

```python
# Extracting Year
df["Year"]=df["InvoiceDate"].dt.year
```

```python
# Extracting Month Number
df["Month"]=df["InvoiceDate"].dt.month
```

```python
# Extracting Month Name
df["Month_Name"]=df["InvoiceDate"].dt.month_name()
```

#### Saving the Clean Dataset

```python
df.to_csv("data/cleaned/online_retail_cleaned.csv",index=False)
```

#### Cleaning Summary

The dataset was cleaned and prepared for exploratory data analysis by performing the following steps:

- Removed duplicate records.
- Removed cancelled transactions.
- Removed records with negative quantities.
- Removed records with invalid (zero or negative) prices.
- Removed records with missing product descriptions.
- Created new features, including **Revenue**, **Month**, **Month_Name**, and **Year**.
- Exported the cleaned dataset for use in the Exploratory Data Analysis (EDA) notebook.

The cleaned dataset is now ready for further analysis and visualization.
